# Making the Plots descripitve

```python
# adding the titles
# adding the labels
# adding a legend
```

```python
plt.plot(df.physicians,df.income,color ='green')
plt.title("Physicians Vs Income")

plt.show()
```

```python
plt.plot(df.work_force,df.income,color ="red",label ="WorkForce")
plt.plot(df.physicians,df.income,color ="green",label ="Physicians")
plt.legend()
plt.title("WorkForce Vs Physicians")

plt.show()
```

```python
plt.scatter(df.work_force,df.income,color ="Red")
plt.title("Scatter Plot")
plt.xlabel("Work_force")
plt.ylabel("Income")
plt.show()
```

# Formating the Plots

```python
plt.plot(df.physicians,df.income,color ="black",label ="physicans")
plt.plot(df.work_force,df.income,color ="#704241",label ="work_force")
plt.plot(df.hospital_beds ,df.income,color  ="c",label ="hospital_beds")
plt.legend()
plt.title("Three color Codes graphes")
plt.show()
```

```python
# adding the markers
plt.plot(df.physicians,df.income,color ="black",label ="physicans",marker ="^") # Triangular marker
plt.plot(df.work_force,df.income,color ="#704241",label ="work_force",marker ="o")# circular marker
plt.plot(df.hospital_beds ,df.income,color  ="c",label ="hospital_beds",marker="x")# Cross marker
plt.legend()
plt.title("Three color Codes graphes")
plt.show()
```

# changing the Plot Size:

```python
plt.figure(figsize=(12,6))
plt.plot(df.hospital_beds,df.income,linestyle ="dotted",color ="red")
plt.title("RGV")
plt.ylabel("income ")
plt.xlabel("Hospital_beds")
plt.show()
```

# Mutliple Plots

```python
plt.subplot(2,1,1)# row ,coloumn, index
plt.plot(df.work_force,df.income ,"^r")
#plt.title("income vs physicians")
plt.show()

plt.subplot(2,1,2)# row , column, index
plt.plot(df.work_force,df.income,"og")
#plt.tilte("work_force vs income")
plt.show()
plt.suptitle("Multpiple Plots")
```

```python
plt.subplot(1,2,1)# row ,coloumn, index
plt.plot(df.work_force,df.income ,"^r")
#plt.title("income vs physicians")
plt.show()

plt.subplot(1,2,2)# row , column, index
plt.plot(df.hospital_beds,df.income,"og")
#plt.tilte("work_force vs income")
plt.show()
plt.suptitle("Multpiple Plots")
```

```python
fig, ax =plt.subplots(nrows =2,ncols =2,figsize=(6,6))# creating the grid of 2*2 matrix with the 6*6 figure
ax[0,0].plot(df.work_force,df.income, "^r")
ax[0,1].plot(df.work_force,df.income, "^g")
ax[1,0].plot(df.hospital_beds,df.income,"ob")
ax[1,1].plot(df.hospital_beds,df.income, "oy")
plt.show()
```
